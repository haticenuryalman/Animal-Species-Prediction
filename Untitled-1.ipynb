```python
import os
import shutil
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

```

```python


# Dataset'inizin bulunduğu ana klasör
dataset_dir = '/Users/haticenuryalman/pc/dataset'

# Eğitim ve test klasörlerinin oluşturulacağı yer
train_dir = '/Users/haticenuryalman/pc/dataset/train'
test_dir = '/Users/haticenuryalman/pc/dataset/test'

# Sınıf isimlerini almak
class_names = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]

# Eğitim ve test için gerekli klasörleri oluşturuyoruz
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Her sınıf için eğitim ve test verilerini ayırma
for class_name in class_names:
    class_path = os.path.join(dataset_dir, class_name)

    if os.path.isdir(class_path):  # Sadece klasörleri işle
        # Bu sınıftaki tüm dosyaları al
        images = [img for img in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, img)) and img.lower().endswith(('jpg', 'jpeg', 'png', 'gif'))]

        # Eğitim ve test verilerini ayırma
        if images:  # Eğer görüntü listesi boş değilse
            train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
        else:
            train_images, test_images = [], []

        # Eğitim için klasör oluştur
        train_class_dir = os.path.join(train_dir, class_name)
        os.makedirs(train_class_dir, exist_ok=True)

        # Test için klasör oluştur
        test_class_dir = os.path.join(test_dir, class_name)
        os.makedirs(test_class_dir, exist_ok=True)

        # Eğitim verilerini kopyala
        for img in train_images:
            shutil.copy(os.path.join(class_path, img), os.path.join(train_class_dir, img))

        # Test verilerini kopyala
        for img in test_images:
            shutil.copy(os.path.join(class_path, img), os.path.join(test_class_dir, img))

print("Veriler başarıyla eğitim ve test setlerine ayrıldı!")
```

```python
import torch.nn as nn
import torch.optim as optim

class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN, self).__init__()
        
        # İlk konvolüsyonel katman (giriş: 3 kanal, çıkış: 32 kanal, kernel boyutu: 3)
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        
        # Max pooling katmanı
        self.pool = nn.MaxPool2d(2, 2)
        
        # Tam bağlantılı katmanlar
        self.fc1 = nn.Linear(128 * 16 * 16, 512)  # 128 kanal, 16x16 boyutunda çıktı
        self.fc2 = nn.Linear(512, num_classes)  # Sonuçları sınıflara ayıran katman

    def forward(self, x):
        # Konvolüsyonel katmanlar ve max pooling
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = self.pool(torch.relu(self.conv3(x)))
        
        # Düzleştirme işlemi (flatten)
        x = x.view(-1, 128 * 16 * 16)
        
        # Tam bağlantılı katmanlar
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        
        return x

# Modeli oluşturuyoruz
model = SimpleCNN(num_classes=len(class_names))

# Cihaz (GPU/CPU) belirleme
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
# Kayıp fonksiyonu (CrossEntropyLoss) ve optimizer (Adam) belirliyoruz
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Eğitim fonksiyonu
def train_model(model, train_loader, criterion, optimizer, num_epochs=10):
    model.train()  # Eğitim moduna alıyoruz
    for epoch in range(num_epochs):
        running_loss = 0.0
        correct = 0
        total = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            
            optimizer.zero_grad()  # Gradients sıfırlanıyor
            
            outputs = model(images)  # Model çıktısı
            loss = criterion(outputs, labels)  # Kayıp hesaplanıyor
            loss.backward()  # Backpropagation ile gradyan hesaplama
            optimizer.step()  # Parametre güncellemesi
            
            running_loss += loss.item()
            
            # Doğru tahmin sayısını hesaplıyoruz
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
        
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}, Accuracy: {100 * correct/total:.2f}%")

# Modeli eğitiyoruz
train_model(model, train_loader, criterion, optimizer, num_epochs=10)
```

```python
# Test fonksiyonu
def test_model(model, test_loader):
    model.eval()  # Test moduna alıyoruz
    correct = 0
    total = 0
    with torch.no_grad():  # Gradients hesaplanmasın
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    accuracy = 100 * correct / total
    print(f"Test Accuracy: {accuracy:.2f}%")

# Modeli test ediyoruz
test_model(model, test_loader)
```

```python
def train_model(model, train_loader, criterion, optimizer, num_epochs=10):
    model.train()
    loss_values = []
    acc_values = []

    for epoch in range(num_epochs):
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        epoch_loss = running_loss / len(train_loader)
        epoch_acc = 100 * correct / total

        loss_values.append(epoch_loss)
        acc_values.append(epoch_acc)

        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")

    return loss_values, acc_values
```

```python
losses, accuracies = train_model(model, train_loader, criterion, optimizer, num_epochs=10)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Loss grafiği
plt.subplot(1, 2, 1)
plt.plot(losses, label='Training Loss', color='red')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Eğitim Kaybı (Loss)')
plt.legend()

# Accuracy grafiği
plt.subplot(1, 2, 2)
plt.plot(accuracies, label='Training Accuracy', color='green')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Eğitim Doğruluğu (Accuracy)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def predict_image(image_path, model, transform, class_names):
    model.eval()

    # Görüntüyü yükle
    image = Image.open(image_path).convert('RGB')
    image_tensor = transform(image).unsqueeze(0).to(device)  # (1, 3, 128, 128)

    # Tahmin yap
    with torch.no_grad():
        outputs = model(image_tensor)
        _, predicted = torch.max(outputs, 1)
        predicted_class = class_names[predicted.item()]
        confidence = F.softmax(outputs, dim=1)[0][predicted.item()].item()

    print(f"Tahmin edilen sınıf: {predicted_class} ({confidence*100:.2f}%)")

    # Görseli göster
    plt.imshow(image)
    plt.title(f"Tahmin: {predicted_class} ({confidence*100:.2f}%)")
    plt.axis('off')
    plt.show()
```
